[magic command] (https://steemit.com/kr/@radajin/how-to-use-jupyter-notebook)

사용 방법
- %(command) : 한줄을 매직커멘드로 사용할때 사용
- %%(command) : 여러줄을 매직커멘드로 사용할때 사용
- %(command)? : 매직키워드 뒤에 ? 를 붙여서 실행하면 키워드에 대한 상세 설명이 나옴

주요 Magic Command
- pwd : 현재 폴더
- ls : 현재 디렉토리 파일 조회
- whos : 현재 정의된 변수 확인
- reset : 현재 정의된 변수 삭제
- time : 실행시간을 알려줌
- timeit : 평균실행 시간을 알려줌
- writefile : 파일작성
- magic : 각 키워드 상세 도움말 출력
- lsmagic : magic 키워드 출력
- run : python file code 실행
- history : 사용한 명령을 출력
- matplotlib inline : matplotlib을 내부에서 실행할수 있도록 함

```python
from keras.datasets import mnist
(train_images, train_labels), (test_images,test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
train_images[0]
```

```python
import matplotlib.pyplot as plt
digit = train_images[4]
plt.imshow(digit, cmap=plt.cm.binary)
```

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype('float32')/255

test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype('float32')/255
```

```python
print(len(train_images[0]))
train_images.shape
```

```python
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels[0]
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images,test_labels)
```

```python
# test 세트의 정확도는 97.8%
test_acc
```
