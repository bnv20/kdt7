```python
import urllib.request as req
import matplotlib.pyplot as plt
import cv2

url = 'https://image.shutterstock.com/z/stock-photo-happy-cheerful-young-woman-wearing-her-red-hair-\
in-bun-rejoicing-at-positive-news-or-birthday-gift-613759379.jpg'

req.urlretrieve(url, 'lady.png')

img = cv2.imread('lady.png')

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) # OpenCV는 BGR 포멧, Matplotlib은 RGB 포멧
plt.show()
```

```python
url = 'http://uta.pw/shodou/img/28/214.png'

req.urlretrieve(url, 'uta.png')

img = cv2.imread('uta.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #OpenCV는 BGR 포맷, Matplotlib은 RGB 포맷
plt.show()
```

```python
import numpy as np
img = np.zeros((512,512,3),np.uint8)
plt.imshow(img)
plt.show()
```

```python
img = cv2.line(img,(0,0),(511,511),(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
img = cv2.rectangle(img,(400,0),(510,128),(0,255,0),5)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img,(450,50),50,(0,0,255),-1)
plt.imshow(img)
plt.show()
```

```python
img = cv2.circle(img,(50,450),50,(0,255,255),2)
plt.imshow(img)
plt.show()
```

- 컬러 매핑 : 주로 그레이 스케일, 투루 컬러 이미지를 많이 활용
- RGB : 빛을 섞을수록 밝아짐(가산혼합), 모니터, TV, 빔프로젝트 등에 사용
- CYMK : 섞을수록 어두어지는 방식(감산혼합), 청록색, 자홍색, 노랑색, 검은색
- YUV : Y축은 밝기 성분을 U,V 두축을 이용하여 색상 표현. 아날로그 방식
- YCbCr : Digital TV에서 사용하는 색공간
